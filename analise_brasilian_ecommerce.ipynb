#### Quais são as categorias com maior e menor receita dos últimos 12 meses?

- Juntando os DataFrames 'order_items' e 'orders' por order_id e depois juntando os mesmos com 'products'

```python
cat_ultimos12_merge = pd.merge(order_items,orders, on=["order_id"], how="inner")
```

```python
cat_ultimos12_merge = pd.merge(cat_ultimos12_merge, products, on=["product_id"], how="inner")
```

```python
# Apenas as colunas necessárias

cat_ultimos12_merge.loc[:, ['order_id', 'order_approved_at', 'price', 'product_category_name_english']].head()
```

```python
# Filtrando os últimos 12 meses

ultimos_12_meses = cat_ultimos12_merge[pd.to_datetime(cat_ultimos12_merge["order_approved_at"]) >= pd.to_datetime(cat_ultimos12_merge["order_approved_at"]).max() - pd.DateOffset(months=12)]
```

```python
# Agrupando por categoria e somando os preços

soma_por_categoria = ultimos_12_meses.groupby("product_category_name_english")["price"].sum()
```

```python
# Encontrar a categoria com a menor receita

categoria_menor_receita = soma_por_categoria.idxmin()
menor_receita = soma_por_categoria.min()
```

```python
print(f"Categoria com maior receita nos últimos 12 meses: {categoria_maior_receita} - Receita: {maior_receita}")
print(f"Categoria com menor receita nos últimos 12 meses: {categoria_menor_receita} - Receita: {menor_receita}")
```

- Gráfico

```python
categorias_ordenadas = soma_por_categoria.sort_values(ascending=False)
categorias_ordenadas = categorias_ordenadas[categorias_ordenadas > 0]
maior_receita_categoria = categorias_ordenadas.idxmax()
menor_receita_categoria = categorias_ordenadas.idxmin()
df_plot = pd.DataFrame({"Categoria": [maior_receita_categoria, menor_receita_categoria], "Receita": [categorias_ordenadas[maior_receita_categoria], categorias_ordenadas[menor_receita_categoria]]})
```

#### Existem sellers que vendem o mesmo produto? Se sim, quais são? Qual é a variação de preço praticado entre os Sellers?

- Juntando os DataFrames 'order_items' e 'products' por product_id

##### Para vendedores que vendem o mesmo produto:

```python
variacao_sellers = pd.merge(order_items, products, on=["product_id"], how="inner")
```

```python
# Identificando os vendedores que vendem o mesmo produto

sellers_repetidos = variacao_sellers.groupby('product_category_name_english')['seller_id'].unique()
```

```python
# Filtrando apenas os produtos com mais de um vendedor

sellers_repetidos = sellers_repetidos[sellers_repetidos.apply(len) > 1]
```

```python
# Exibir os vendedores distintos que vendem o mesmo produto

print(f"Vendedores distintos que vendem o mesmo produto: {sellers_repetidos}")
```

##### Para variação de preço praticado:

```python
# Agrupar os dados pelo campo 'product_id'

preco_praticado = variacao_sellers.groupby('product_id')
```

```python
# listas para armazenar os dados

produtos = []
variacao_precos = []
```

```python
# Iterar sobre cada grupo

for produto, grupo in preco_praticado:
    if len(grupo['seller_id'].unique()) > 1:
        variacao_preco = grupo['price'].max() - grupo['price'].min()
        produtos.append(produto)
        variacao_precos.append(variacao_preco)
```

```python
# DataFrame com os dados

resultado_variacao_preco = pd.DataFrame({'Produto': produtos, 'Variação de Preço': variacao_precos})
```

```python
print(f"Variação de preço praticado entre os vendedores para os mesmos produtos: {resultado_variacao_preco}")
```

- Gráfico da variação de preço praticado

```python
resultado_variacao_preco_sorted = resultado_variacao_preco.sort_values(by='Variação de Preço', ascending=False)

plt.figure(figsize=(10, 4))
plt.scatter(range(len(resultado_variacao_preco_sorted)), resultado_variacao_preco_sorted['Variação de Preço'], color='#72BCD4', alpha=0.5)
plt.xlabel('Produto')
plt.ylabel('Variação de Preço')
plt.title('Variação de Preço entre Vendedores para os Mesmos Produtos')
plt.tight_layout()
plt.show()
```

#### Ouve inflação no preço dos produtos ao passar do tempo? Se ouve, de quanto foi a variação desta inflação em % e em R$?

```python
# Extrair o ano e o mês da coluna de datas

order_items['ano'] = order_items['shipping_limit_date'].dt.year
order_items['mes'] = order_items['shipping_limit_date'].dt.month
```

```python
# Agrupar os dados por ano e calcular o preço médio de todos os produtos para cada ano

preco_medio_ano = order_items.groupby('ano')['price'].mean()
```

```python
# Calcular a variação de preço em reais entre os anos consecutivos

variacao_preco_reais = preco_medio_ano.diff()
```

```python
# Calcular a variação de preço em porcentagem entre os anos consecutivos

variacao_preco_percentual = (preco_medio_ano.pct_change() * 100).round(2)
```

```python
print(f"Variação de preço médio por ano: \n{preco_medio_ano}")
print(f"\nVariação de preço em reais entre os anos consecutivos: \n{variacao_preco_reais}")
print(f"\nVariação de preço em porcentagem entre os anos consecutivos: \n{variacao_preco_percentual}")
```

- Preenchendo os valores NaN com 0, para não serem excluidos. Isso indica que não houve variação de preço em relação ao ano anterior devido à falta de dados.

```python
variacao_preco_reais = preco_medio_ano.diff().fillna(0)
variacao_preco_percentual = (preco_medio_ano.pct_change() * 100).fillna(0).round(2)
```

```python
print(f"Variação de preço médio por ano: \n{preco_medio_ano}")
print(f"\nVariação de preço em reais entre os anos consecutivos: \n{variacao_preco_reais}")
print(f"\nVariação de preço em porcentagem entre os anos consecutivos: \n{variacao_preco_percentual}")
```

- Gráfico

```python
plt.figure(figsize=(12, 4))

# Gráfico de linha para a variação de preço em reais entre os anos consecutivos
plt.subplot(1, 2, 1)
variacao_preco_reais.plot(marker='o', color='#72BCD4')
plt.xlabel('Ano')
plt.ylabel('Variação de Preço (R$)')
plt.title('Variação de Preço em reais entre os Anos Consecutivos')


# Gráfico de linha para a variação de preço em porcentagem entre os anos consecutivos
plt.subplot(1, 2, 2)
variacao_preco_percentual.plot(marker='o', color='green')
plt.xlabel('Ano')
plt.ylabel('Variação de Preço (%)')
plt.title('Variação de Preço em Porcentagem entre os Anos Consecutivos')

# Ajustar o layout e exibir os gráficos
plt.tight_layout()
plt.show()
```

#### Existe Relação entre a quantidade de vendas e a quantidade de reviews para os sellers? É possível identificar aumento ou queda na venda de um sellers com base nas avaliações do que já foi vendido no passado?

- Juntando os DataFrames 'order_items' e 'order_reviews_preenchida' por order_id

```python
merge_dados = pd.merge(order_items, order_reviews_preenchida, on='order_id', how='inner')
merge_dados = pd.merge(merge_dados, sellers, on='seller_id', how='inner')
```

```python
# Calculando a média das avaliações anteriores por seller_id

reviews_mean = merge_dados.groupby('seller_id')['review_score'].mean().reset_index()
reviews_mean.columns = ['seller_id', 'avg_review_score']
```

```python
# Calculando a média das vendas anteriores por seller_id

sales_mean = merge_dados.groupby('seller_id')['order_id'].count().reset_index()
sales_mean.columns = ['seller_id', 'avg_sales']
```

```python
# Merge dos dados de média de vendas e média de avaliações

merged_means = pd.merge(reviews_mean, sales_mean, on='seller_id')
```

```python
print(merged_means)
```

```python
plt.figure(figsize=(10, 4))
plt.bar(merged_means.index, merged_means['avg_sales'], alpha=0.5)
plt.title('Média de Vendas Anteriores por Seller')
plt.xlabel('Seller')
plt.ylabel('Média de Vendas Anteriores')
plt.grid(True)
plt.show()
```

- Não parece haver uma relação significativa entre as avaliações e as vendas dos sellers.

```python
# Adicionando uma constante para a regressão linear

x = sm.add_constant(merged_means['avg_review_score'])
y = merged_means['avg_sales']
```

```python
# Ajustando o modelo de regressão linear

model = sm.OLS(y, x).fit()
```

```python
print(model.summary())
```
